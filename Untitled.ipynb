```python
import requests
from bs4 import BeautifulSoup
from collections import defaultdict

USER_AGENT = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64;AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
response = requests.get('https://datascience.stackexchange.com/tags?page=1&tab=name', headers=USER_AGENT,stream=True)
bea_obj=BeautifulSoup(response.content)
pages=bea_obj.find_all('span',{'class':'page-numbers'})
tags=defaultdict(list)
page_no=[]
for page_tag in pages:
    try:
        page_no.append(int(page_tag.get_text()))
    except ValueError:
        pass
for i in range(2):
    response = requests.get('https://datascience.stackexchange.com/tags?page='+repr(i+1)+'&tab=name', headers=USER_AGENT,stream=True)
    bea_obj=BeautifulSoup(response.content)
    for tag in bea_obj.find_all('a',{'class':"post-tag"}):
        tags[tag.get_text()]={'link':'https://datascience.stackexchange.com'+tag['href']}
        response_ques= requests.get('https://datascience.stackexchange.com'+tag['href']+'?sort=newest&days=7', headers=USER_AGENT,stream=True)
        bea_obj_ques=BeautifulSoup(response_ques.content)
        print('https://datascience.stackexchange.com'+tag['href']+'?sort=newest&days=1')
        tags[tag.get_text()]['questions']=len(bea_obj_ques.find_all('a',{'class':"question-hyperlink"}))
        response_ques= requests.get('https://datascience.stackexchange.com//tags/'+tag.get_text()+'/info', headers=USER_AGENT,stream=True)
        bea_obj_ques=BeautifulSoup(response_ques.content)
        print('https://datascience.stackexchange.com/tags/'+tag.get_text()+'/info')
        print(bea_obj_ques.find_all('b'))
        tags[tag.get_text()]['views']=[view.get_text() for view in (bea_obj_ques.find_all('b')) if (view.get_text()).endswith('times')]
```
