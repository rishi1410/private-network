# Search Engine:
A search engine is the practical application of information retrieval technique. Search engine is used in many applications like desktop,websites etc. A Web Search Engine crawles through the websites to find all the web pages and scrape keywords corresponding to that webpage and stores webpage url and keywords in its database. 

# Web search engine:
A web search engine database maintains an index and short summaries of billions of pages on the Web. Yahoo was originally known as a "directory" rather than a search engine, because it indexed much of its content by human observation. However, as Web content grew exponentially, it became impossible to index everything manually.So this process was automated by using Web crawlers.<br>
**Types of Web Search Engine:**<br>
The Web has become a major source of information source in the last decade. Based on how pages are being indexed and query are generated are handled by search engines, They are divided into four types:
* Crawler-based search engines
* Human powered directories
* Hybrid Search engine
* Meta Search engine

## Crawler-based Search Engine:
 Crawler-based search engine is a automated indexing system which uses web crawlers to get all the urls and keywords from the webpage and saves these urls and their associated keywords in the database. When a query is made to a search engine these database is accessed to send matching webpages to the user.
## Human powered directories:
Directories are a collection of web sites that are organized into broad categories. Yahoo was originally a directory based search engine. When you search a directory, you are only searching the web sites identified and included in that directory
## Hybrid Search Engine:
Hybrid Search Engines is combination of crawler based and directories for indexing webpages. Most of Crawler-based and directories based search engines are moving towards Hybrid search engine.
## Meta Search Engine:
Meta Search Engine scrapes results from other search engines and feeds them to its own page ranking algorithm. Then showcases the results in the new page rank order. 
## Architecture of a Web Search Engine:
Two major functions of a search engine are Index process and Query process.Index Process is a continuous process with uses Webcrawler to scrape information from web and saves the results in a database.Query process uses these database and the query generated by the user to provide results in which are sorted by page ranking algorithm.

##### Crawler:
In many applications, the crawler component has the primary responsibility for identifying and acquiring documents for the search engine.A web crawler can be restricted to a single site, such as a university, as the basis for site search. Focused, or topical, web crawlers use classification techniques to restrict the pages that are visited to those that are likely to be about a specific topic. This type of crawler may be used by a vertical or topical search application, such as a search engine that provides access to medical information on web pages.

```python
# crawler that gets URL Document from the web
import requests
USER_AGENT = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64;AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
response_data = requests.get('https://refactored.ai/', headers=USER_AGENT,stream=True)
print(response_data)
```

'requests' is a HTTP python library that helps us to make HTTP request from python program.User_Agent is the variable that is used to specify requesting users system configuration and browser name.In requests.get function we are  sending a 'get' to https://refactored.ai/ server acting like a Chrome browser whose version is '61.0.163.100' on windows 10 64bit system. get is method of requests module to request webpage information at location https://refactored.ai/. This method returns a reponse object which we stored in response_data variable. We can get status of the response by using status_code attribute  of the response object. This attribute returns a int value of the response code.we save the content,headers and status_code of the response object to database and send  a copy of this data to conversion block. 

```python
print('response Code:'+repr(response_data.status_code))
print('response header:'+repr(response_data.headers))
```

##### Conversion:
We are going to use Content-Type key of response object header to know the type of file we are dealing with and character decoding method to implement in order to get readable text that can be processed by text transformation block to get all the keywords in the file. From above we can see that we are dealing with a html file which used UTF-8 character encoding method to encode the text that we requested.Now we have raw content of the data and it's character encoding method and file type.From below we can see raw content of the response looks like big chunk of a pair of hexadecimal values. Now we are going to decode  this data using UTF-8.

```python
decode_method=response_data.encoding
print(decode_method)
print('Byte Response of the object:'+repr(response_data.content.hex()))
```

We decoded content that can be decoded by UTF-8 Character decoding method and leaving the bytes that cannot be converted which will not be the case here as encoding method used here was also UTF-8

```python
html_file=response_data.content.decode(decode_method, "ignore")
print(html_file)
```

### Text Acquisition:
Now we are going to get all the urls that are present in this webpage. We are going to use BeautifulSoup package to prase our html file that we have downloaded by going to https://refactored.ai/. We are going to use find_all fuction attribute of BeautifulSoup object to find all the tags that have attribute 'href'. We are converting our html response into BeautifulSoup object.

```python
from bs4 import BeautifulSoup
beaut_obj=BeautifulSoup(html_file)
Tags_href=(beaut_obj.find_all(href=True))
Tags_href
```

We now can create a dictionary with internal and external as keys with internal links and external links as their values.We have Tags_href list that contains all the tags  that have href attribute we are going get each element of the list using a for loop.
Then we are going to discard all tags that have href attribute value as a string punctuations.These filtered out tags are divided into internal link if href attribute value starts with '/' character. Now we have all the links in the web page.

```python
from collections import defaultdict
import string
href_links=defaultdict(set)
href_links['base']='https://refactored.ai/'
for tag_href in Tags_href:
    if (tag_href['href'] not in string.punctuation):
        if tag_href['href'].startswith('/'):
            href_links['internal'].add(tag_href['href'])
        else:
            href_links['external'].add(tag_href['href'])
href_links   
```

Now we are going to extract title,meta description of the website and content of the website.We get title using find method and passing title as the argument.Meta description gives us the overview of the website. To get all the content present in the website we can use function object get_text() of the BeautifulSoup object.

```python
content=beaut_obj.get_text()
title=(beaut_obj.find('title')).content
summary=(beaut_obj.find('meta',{'name':"description"}))['content']
print('Title of webpage:'+repr(title))
print('Summary of webpage:'+repr(summary))
print('content:'+repr(content))
```

Now we are going to clean the content by removing punctuations,blank lines and numbers.In the following code we have loaded regular expression library so that we can replace patterns in the content by blank space.re.sub method is used to replace a matting substring of the string with a blank space.\w\s tells us to match all the substrings that have alphanumeric characters followed by a white space.\[ ^\\w\\s\]negates the above statement.\\n+ means a any number of new lines. \[0-9\] matches any numeric value.'+' is used to represent multiples repetitions of the preceding character 

```python
import re
content = re.sub(r'[^\w\s]',' ',content) 
content = re.sub('\n+'," ", content)
content = re.sub('\[[0-9]+\]'," ",content)
content = re.sub('[0-9]+'," ",content)
content = re.sub(' +'," ",content)
content
```

### Index Creation:
Now we are going to convert content into unigrams. We have called split function object of sting object to split the string into substring at every whitespace. Now we are going to use itertools library to coubt number of times a keyword appears in the content.

```python
from collections import Counter
keyword_list=[]
content_list=content.split(' ')
commonWords = ["the", "be", "and", "of", "a", "in", "to", "have", "it","i", "that", "for", "you", "he", "with", "on", "do", "say", "this", "they", "is", "an", "at", "but","we", "his", "from", "that", "not","by", "she", "or", "as", "what", "go", "their","can", "who", "get","if", "would", "her", "all", "my", "make", "about", "know", "will","as", "up", "one", "time", "has", "been", "there", "year", "so","think", "when", "which", "them", "some", "me", "people", "take",
        "out", "into", "just", "see", "him", "your", "come", "could", "now","than", "like", "other", "how", "then", "its","our", "two","was","more","these", "want","way", "look", "first", "also", "new", "because","day", "more","use", "no", "man", "find", "here", "thing", "give","many", "well","are","same"]    
for word in content_list:
    if word.lower() not in commonWords:
        keyword_list.append(word)
count_keyword=Counter(keyword_list)
count_keyword
```
